#Chart Requirements

```python
# Monthly Loan Application Trend – Area Chart

monthly_trend = (
    df
    .groupby(df['issue_date'].dt.month_name())['id']
    .count()
    .reindex([
        'January','February','March','April','May','June',
        'July','August','September','October','November','December'
    ])
    .reset_index(name='Total Applications')
)

import matplotlib.pyplot as plt

plt.figure(figsize=(7,3))

plt.fill_between(
    monthly_trend['issue_date'],
    monthly_trend['Total Applications'],
    alpha=0.6
)

plt.plot(
    monthly_trend['issue_date'],
    monthly_trend['Total Applications'],
    marker='o'
)

# ✅ Correct value labels
for i in range(len(monthly_trend)):
    plt.text(
        monthly_trend['issue_date'][i],
        monthly_trend['Total Applications'][i],
        f"{monthly_trend['Total Applications'][i]:,.0f}",
        ha='center',
        fontsize=7
    )

plt.title('Monthly Loan Application Trend (Area Chart)')
plt.xlabel('Month')
plt.ylabel('Total Applications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#**Regional Analysis by State (Bar Chart)**

```python
Reginol_Analysis = df.groupby('address_state')['id'].count() .sort_values(ascending=False)
plt.figure(figsize=(7,3))
plt.bar(Reginol_Analysis.index, Reginol_Analysis.values , color='skyblue')


for i in range(len(Reginol_Analysis)):
  plt.text(Reginol_Analysis.index[i],Reginol_Analysis.values[i], f"{Reginol_Analysis.values[i]}",ha='center')


plt.title('Regional Analysis by State')
plt.xlabel('State')
plt.ylabel('Total Applications')
plt.xticks(rotation=-90)
plt.tight_layout()
plt.show()

```

```python
Regional_Analysis = (
    df
    .groupby('address_state')['id']
    .count()
    .sort_values(ascending=True)   # ascending for horizontal
)

import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.barh(
    Regional_Analysis.index,
    Regional_Analysis.values
)

# Value labels
for i in range(len(Regional_Analysis)):
    plt.text(
        Regional_Analysis.values[i] + 50,   # small offset
        Regional_Analysis.index[i],
        f"{Regional_Analysis.values[i]:,}",
        va='center',
        fontsize=7
    )

plt.title('Regional Analysis by State')
plt.xlabel('Total Applications')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
#Loan Term Analysis (Donut Chart):

loanterm_Analysis = df['term'].value_counts()
plt.figure(figsize=(7,3))
plt.pie(loanterm_Analysis, labels=loanterm_Analysis.index, autopct='%1.1f%%', startangle=90)
plt.title('Loan Term Analysis')
plt.tight_layout()
plt.show()
```

```python
#Employee Length Analysis (Bar Chart)
emp_length = df['emp_length'].value_counts().sort_index()
plt.figure(figsize=(7,3))
plt.bar(emp_length.index , emp_length.values , color='orange', label='Employee Length Analysis')

for i in range(len(emp_length)) :
  plt.text(i, emp_length.values[i] , f"{emp_length.values[i]:,}" , ha='center' , va='bottom' , fontsize=7)

plt.title('Employee Length Analysis')
plt.xlabel('Employee Length')
plt.ylabel('Total Applications')
plt.tight_layout()
plt.legend()
plt.xticks(rotation=-90)
plt.show()
```

```python
#  Loan Purpose Breakdown (Bar Chart)

loan_purpose = df['purpose'].value_counts()

plt.figure(figsize=(7,3))
plt.bar(loan_purpose.index , loan_purpose.values , color='g'  , label='Loan Purpose Breakdown' )

for i in range(len(emp_length)) :
  plt.text(i, loan_purpose.values[i] , f"{loan_purpose.values[i]:,}" , ha='center' , va='bottom' , fontsize=7 , color='purple')

plt.title('Loan Purpose Breakdown')
plt.xlabel('Purpose')
plt.ylabel('Total Application')
plt.tight_layout()
plt.legend()
plt.xticks(rotation=-90)
plt.show()


```

```python
# Home Ownership Analysis (Tree/Heat Map)

heatmap_data = pd.crosstab(
    df['home_ownership'],
    df['loan_status']
)

plt.figure(figsize=(7,4))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=',',
    cmap='Reds'
)

plt.title('Home Ownership vs Loan Status (Heat Map)')
plt.xlabel('Loan Status')
plt.ylabel('Home Ownership')
plt.tight_layout()
plt.show()

```

```python


# Monthly Trends by Issue Date (Line / Area Chart)

monthly_trend = (
    df
    .groupby(df['issue_date'].dt.month_name())['loan_amount']
    .sum()
    .divide(1000000)

    .reset_index(name='Total Funded Amount (Million)')
)

print(monthly_trend)



plt.figure(figsize=(7,3))

plt.plot(
    monthly_trend['issue_date'],
    monthly_trend['Total Funded Amount (Million)'],
    marker='o',
    markersize=6,
    label='Monthly Funded Amount'
)

# Value labels
for i in range(len(monthly_trend)):
    plt.text(
        monthly_trend['issue_date'][i],
        monthly_trend['Total Funded Amount (Million)'][i],
        f"{monthly_trend['Total Funded Amount (Million)'][i]:,.1f}",
        ha='center',
        fontsize=7
    )

plt.title('Monthly Funded Amount Trend by Issue Date')
plt.xlabel('Month')
plt.ylabel('Total Funded Amount (Million)')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()





```
